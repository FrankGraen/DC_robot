```python


from roboticstoolbox import *
from spatialmath import *
from math import pi
import numpy as np
```

```python
DFbot = DHRobot(
    [
                    RevoluteDH(alpha=np.pi/2,qlim=np.array([-np.pi,np.pi])),
                    RevoluteDH(a=0.15,qlim=np.array([-np.pi,np.pi])),
                    RevoluteDH(a=0.15,qlim=np.array([-np.pi,np.pi])),
                    RevoluteDH(d=-0.05494,alpha=np.pi/2,offset=np.pi/2,qlim=np.array([-np.pi,np.pi])),
                    RevoluteDH(d=0.068,alpha=-np.pi/2,qlim=np.array([-np.pi,3/2*np.pi])),  
                    RevoluteDH(d=0.033,qlim=np.array([-np.pi,3/2*np.pi]))

 
    ],
    name="DFbot",
)
```

```python
state0=[0,0,0,0,0,0]
state1=[0,np.pi/2,-np.pi/2,0,0,0]
```

```python
#笛卡尔运动
```

```python
state0=[0,0,0,0,0,0]
T0=DFbot.fkine(state0)
state1=[0,np.pi/2,-np.pi/2,0,0,0]
T1=DFbot.fkine(state1)
qt = rtb.tools.trajectory.ctraj(T0,T1, 10)
#DFbot.plot(qt, backend='pyplot', movie='panda3.gif')
```

```python
qt.plot()
```

```python
qt
```

```python
pose_list=[]
for i in range(9):
    sol = DFbot.ikine_LM(qt[i+1],q0=[0,0,0,0,0,0],ilimit=100, slimit=100,joint_limits=True,tol=0.005)
    pose_list.append(sol.q)
```

```python
pose_list
```

```python
# 执行转换
pose_array=np.array(pose_list)
```

#显示机械臂运动

```python
DFbot.plot(pose_array, backend='pyplot', movie='panda3.gif')
```
